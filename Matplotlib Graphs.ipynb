```python
#from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
```

```python
#Line Chart
x = [1,2,3]
y = [4,5,2]
plt.plot(x,y)
```

```python
x=[1,2,3]
y=[4,5,2]
plt.plot(x,y)
plt.title('Line Graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
# Adding style elements to a graph - Plotting multiple graphs
#First Data
x1=[1,2,3]
y1=[4,5,2]

#Second Data
x2=[6,7,8]
y2=[9,10,7]

plt.plot(x1,y1,label='first',color='red',linewidth=5)
plt.plot(x2,y2,label='second',color='m',linewidth=5)

plt.title('Style graph')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.legend()
plt.grid(True,color='lightblue')

plt.show()
```

```python
help(plt.legend)
```

### Bar Plot

```python
# graph that presents categorical data

#Category 1
x1=[1,3,5,7]
y1=[5,2,6,6]

#Category 2
x2=[2,4,6,8]
y2=[6,3,5,2]

plt.bar(x1,y1,label='first')
plt.bar(x2,y2,label='second')

plt.legend()
plt.xlabel('bar numbers')
plt.ylabel('bar height')
plt.title('Bar chart')
plt.show()
```

```python
help(plt.bar)
```

## Histogram Plot

```python
# univariate distribution : used to represent the distribution of data
sale_values=[12,10,5,7,0,14,15,24,10,11,15,30,18,11,10,18,30]

bins = [0,5,10,15,20,25,30]

plt.hist(sale_values,bins,histtype='bar',rwidth=0.6)

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Histogram graph')
plt.show()
```

### Scatter plot

```python
# used to check the correlation among variables
x=[1,3,4,8,5,6,7]
y=[3,5,6,7,8,9,3]

plt.scatter(x,y,color='b')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Scatter plot')
plt.show()
```

### Stack plot

```python
days=[1,2,3,4,5] #x axis

#y axis values
sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]

plt.stackplot(days,sleeping,eating,working,playing,colors=['m','c','r','k'],
              labels=['sleeping','eating','working','playing'])

plt.xlabel('days')
plt.ylabel('time spent')
plt.legend()
plt.title('Daily routine')
plt.show()
```

### Pie chart

```python
#Illustrate numerical proportion of data
Mylabels=('python','c++','ruby','java')
data=[215,130,245,210]

Mycolors=['gold','yellowgreen','lightcoral','lightskyblue']

plt.pie(data,labels=Mylabels,colors=Mycolors,autopct='%1.2f%%')

plt.show()
```
